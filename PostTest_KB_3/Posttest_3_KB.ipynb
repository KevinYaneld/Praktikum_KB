## IMPORT LIBRARY

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

## MEMBACA DATASET

```python
df = pd.read_csv("car_price_prediction.csv")
```

```python
df.head()
```

## VISUALISASI DATA

### GRAFIK PIE

```python
data_pie = df["Drive wheels"].value_counts()
data_pie = df["Drive wheels"].value_counts().rename_axis("Drive wheels")
data_pie = df["Drive wheels"].value_counts().rename_axis("Drive wheels").reset_index(name="ID")
data_pie.head(10)
```

```python
plt.title("Grafik Pie Jenis Mobil Berdasarkan Ban Penggerak")
plt.pie(data_pie["ID"], labels=data_pie["Drive wheels"], startangle=90, autopct="%.2f%%")
plt.show()
```

### GRAFIK BATANG

```python
# Dengan matplotlib
x = list(df["Category"])
y = list(df["Price"])

plt.xlabel("Harga")
plt.ylabel("Kategori")

plt.title("Grafik Batang Harga Dengan Kategori")
plt.barh(x, y, color="red")
plt.show()
```

### GRAFIK SCATTER PLOT

```python
plt.title("Grafik Scatter Plot Perbandingan Harga Dengan Jumlah Airbag")
plt.xlabel("Harga")
plt.ylabel("Airbag")
plt.scatter(x=df["Price"], y=df["Airbags"])
plt.show()
```
